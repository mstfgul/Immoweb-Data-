```python
#isoleted values without data and finding the mean
mean_construction_year = df['construction_year'][df['construction_year'].notnull()].mean()
#int because take the just year
mean_construction_year = int(mean_construction_year)
#drop the wrong values
df = df.drop(df[df['construction_year'] > 2034].index)
#full the nulls with the mean of consyear
df['construction_year'] = df['construction_year'].fillna(mean_construction_year)
```

```python
df.groupby('district')['price'].sum().sort_values(ascending=False)
```

```python
houses_counter_per_district = df.groupby('district').size().sort_values(ascending=False)
print(houses_counter_per_district)
```

```python
houses_counter_per_district = df.groupby('district').size().sort_values(ascending=False).head(10)


plt.figure(figsize=(5, 3))  
houses_counter_per_district.plot(kind='barh', color='skyblue') 
plt.xlabel('HOUSES') 
plt.ylabel('DISTRICTS') 
plt.tight_layout()
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Korelasyon matrisini hesaplayalım
corr_matrix = numeric_columns.corr(method='pearson')

# Korelasyon matrisini heatmap olarak görselleştirelim
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='inferno', fmt='.2f')
plt.show()
```
