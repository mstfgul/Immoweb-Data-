```python
spd = df_hui.groupby('District')['pricesqm'].mean()

# Ordenar los valores de la Serie resultante en orden descendente
spd = spd.sort_values(ascending=False)

# Convertir la Serie a un DataFrame para su uso con seaborn
spd = spd.reset_index()
print(spd)
```

```python
df_hui = df_hui.select_dtypes(include=['float64', 'int'])
corr_matrix = df_hui.corr()
corr_matrix_np = corr_matrix.values
np.fill_diagonal(corr_matrix_np, 1)
corr_matrix = pd.DataFrame(corr_matrix_np, index=corr_matrix.index, columns=corr_matrix.columns)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=False, cmap="viridis", vmax=1, vmin=-1)
plt.title("Correlations")
plt.show()
```
